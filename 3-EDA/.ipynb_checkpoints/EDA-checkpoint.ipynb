<div style=" padding: 30px; text-align: left; color: #000;">
    <h3 style="font-family: 'calibri', sans-serif;ext-align: center;  font-size: 20px; margin-top: 0; ">
    What percentage of loan applicants are under 50 years old?
    </h3>
</div>

```python
len(df[(df['age'] < 50)]) / len(df) * 100
```

<div style=" padding: 30px; text-align: left; color: #000;">
    <h3 style="font-family: 'calibri', sans-serif;ext-align: center;  font-size: 20px; margin-top: 0; ">
    Making "yes", "no" coulumns into "1", "0"
    </h3>
</div>

```python
zero_one = {
    'yes': 1,
    'no': 0
}
yes_no_columns = ['default', 'housing', 'loan', 'y']
for i in yes_no_columns:
    df[i] = df[i].map(zero_one)
df
```

```python
print("Percentage of subscribed clients =", '{:.1%}'.format(df["y"].mean()))
```

```python
print("Average age of single subscribed clients =", int(df[(df["y"] == 1) & (df["marital"] == "single")]["age"].mean()), "years")
```

<div style=" padding: 30px; text-align: left; color: #000;">
    <h3 style="font-family: 'calibri', sans-serif;ext-align: center;  font-size: 20px; margin-top: 0; ">
    Subscription & Marital:
    </h3>
</div>

```python
pd.crosstab(df["y"], df["marital"])
```

<div style=" padding: 30px; text-align: left; color: #000;">
    <h3 style="font-family: 'calibri', sans-serif;ext-align: center;  font-size: 20px; margin-top: 0; ">
    Subscription & Marital(Percentage):
    </h3>
</div>

```python
pd.crosstab(df["y"], df["marital"], normalize = 'index')
```

<div style=" padding: 30px; text-align: left; color: #000;">
    <h3 style="font-family: 'calibri', sans-serif;ext-align: center;  font-size: 20px; margin-top: 0; ">
    The average age and duration based on job title:
    </h3>
</div>

```python
df.pivot_table(["age", "duration"], ["job"], aggfunc = "mean")
```
