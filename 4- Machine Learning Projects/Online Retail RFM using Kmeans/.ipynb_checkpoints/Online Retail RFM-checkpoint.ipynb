# Online Retail RFM using Kmeans

Welcome.

In this notebook we're gonna be looking at a dataframe which is for an online retail, and:
>Try to put related customers in the same group.

As you know, this is a clustering problem.

We're gonna work on it using RFM method

What we're gonna cover here:
* Introduction to RFM method
* Import libraries and data
* Data understanding and preparation
* Create RFM dataset
* Create the model and fit the data

## Introduction to RFM method

In this method, we have 3 columns:
1. **Recency:** The last time a customer bought something.
2. **Frequency:** How many times did a customer order something.
3. **Monetary:** The total mony a costomer spent on our products.

So we need to grab useful data from our dataframe in order to create such a dataset.

## Import libraries and data

```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.cluster import KMeans
```

```python
data = pd.read_excel('Online Retail.xlsx')
data.head()
```

## Data understanding and preparation

```python
data.info()
```

As you might guess, we can not accept rows without a `CustomerID` since the whole point of this problem is to know the customer behavior!

So we need to delete the rows without a `CustomerID`

```python
data = data[data['CustomerID'].notnull()]
```

```python
data.info()
```

Now that looks better :)

The next thing is that we dont need the time in the `InvoiceDate` column. So let's create a new column called `InvoiceDay` in order to only have date without time

```python
data['InvoiceDay'] = data['InvoiceDate'].apply(lambda x: dt.datetime(x.year, x.month, x.day))
```

In order to make **Recency**, we need to set a `pin_date` which would be the MAX date in our dataframe (plus 1) so that each date, would be compared to `pin_date` to understand when was the last time that the customer bought something

```python
pin_date = max(data['InvoiceDay']) + dt.timedelta(1)
pin_date
```

Our dataframe only has Quantity and UnitPrice. But what we want is `TotalSum`.

```python
data['TotalSum'] = data['Quantity'] * data['UnitPrice']
data.head()
```

## Create RFM dataset

Now it's time to create our RFM dataset

```python
rfm = data.groupby('CustomerID').agg({
    'InvoiceDay': lambda x: (pin_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalSum': 'sum'
})
rfm
```

```python
rfm.rename(columns= {
    'InvoiceDay': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSum': 'Monetary'
}, inplace=True)
rfm
```

Now we want to scale them. Instead of using sklearn scalers, let's use the **q-cut** method. It would label the numbers from 1 to q. let's set q to 4 (q=4)

```python
r_labels = range(4, 0, -1) 
r_groups = pd.qcut(rfm['Recency'], q=4, labels=r_labels)
f_labels = range(1, 5) 
f_groups = pd.qcut(rfm['Frequency'], q=4, labels=f_labels)
m_labels = range(1, 5)
m_groups = pd.qcut(rfm['Monetary'], q=4, labels=m_labels)
```

```python
rfm['R'] = r_groups.values
rfm['F'] = f_groups.values
rfm['M'] = m_groups.values
rfm
```

## Create the model and fit the data

```python
X = rfm[['R', 'F', 'M']].values
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
rfm['kmeans_cluster'] = kmeans.labels_
```

```python
rfm
```
